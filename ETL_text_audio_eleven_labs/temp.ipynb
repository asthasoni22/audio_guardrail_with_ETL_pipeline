```python
import pandas as pd
import os

# Path to your Excel file
excel_file = "/home/ubuntu/TTS_elevenlabs/input/voice_ids.xlsx"

# Output folder (will be created if not exists)
output_folder = "voice_ids"
# os.makedirs(output_folder, exist_ok=True)

# Only these sheet names will be processed
categories = [
    "male_adult",
    "male_young",
    "male_old",
    "female_adult",
    "female_young",
    "female_old",
    "joyful",
    "depressed",
    "angry",
    "crying",
    "giggle",
    "whisper",
    "neutral"
]

# Load the Excel file
sheets = pd.read_excel(excel_file, sheet_name=None)

# Process only the required sheets
for sheet_name in categories:
    if sheet_name in sheets:
        df = sheets[sheet_name]

        if "ids" in df.columns:
            ids = df["ids"].dropna().astype(str)

            txt_path = os.path.join(output_folder, f"{sheet_name}.txt")
            with open(txt_path, "w", encoding="utf-8") as f:
                for id_val in ids:
                    f.write(f"{id_val}\n")

            print(f"✅ Created {txt_path} with {len(ids)} ids")
        else:
            print(f"⚠️ Sheet '{sheet_name}' has no 'ids' column, skipped")
    else:
        print(f"⚠️ Sheet '{sheet_name}' not found in Excel, skipped")
```
